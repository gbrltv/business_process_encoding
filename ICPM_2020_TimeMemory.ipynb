```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


df_multiclass = df.loc[pd.isnull(df.f1_binary),:]
df_binary = df.drop(df_multiclass.index)

#df_acc = df_binary['encoding','accuracy', 'f1_binary', 'precision_binary', 'recall_binary']

fig= plt.figure(figsize=(13,12))
ax = sns.violinplot(x="time", y="encoding", data=df_binary)
#ax.set(ylim=(0.85, 1.0))
```

```python
from yellowbrick.features import PCA


X = df_binary[['accuracy', 'mem_rf', 'mem_total', 'time']]

y = df_binary['encoding'].astype('category').cat.codes

fig= plt.figure(figsize=(10,10))
visualizer = PCA(scale=True, proj_features=True, classes=pd.unique(df_binary['encoding']), 
                heatmap=True)
visualizer.fit_transform(X, y)
visualizer.show()
```
